```python
import tkinter as tk

# Function to evaluate the expression
def calculate():
  try:
    result = eval(entry.get())
    entry.delete(0, tk.END)
    entry.insert(tk.END, str(result))
  except Exception as e:
    entry.delete(0, tk.END)
    entry.insert(tk.END, "Error")

# Function to handle button clicks (including equals)
def button_click(value):
  if value == "=":
    calculate()
  else:
    entry.insert(tk.END, value)

# Function to clear the display
def clear_display():
  entry.delete(0, tk.END)

# Create the main application window
root = tk.Tk()
root.title("Simple Calculator")
root.configure(background="black")  # Set background color to black

# Create the text box for input/output
entry = tk.Entry(root, width=40, borderwidth=5, bg="lightgray")  # Light gray entry background
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Define button colors
button_colors = {
  "numbers": "lightblue",  # Light blue for numbers
  "operators": "lightcoral",  # Light coral for operators
  "clear": "lightgray"       # Light gray for clear button
}

# Create buttons for numbers and operators
buttons = [
  ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("+", 1, 3),
  ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("-", 2, 3),
  ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("*", 3, 3),
  ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("/", 4, 3),
  ("C", 5, 0)  # Clear button definition
]

# Create buttons and assign function to handle clicks
for (text, row, col) in buttons:
  if text == "C":
    button = tk.Button(
        root, text=text, padx=20, pady=10, command=clear_display, bg=button_colors["clear"]
    )
  elif text in "+-*/":
    button = tk.Button(
        root, text=text, padx=20, pady=10, command=lambda value=text: button_click(value), bg=button_colors["operators"]
    )
  else:
    button = tk.Button(
        root, text=text, padx=20, pady=10, command=lambda value=text: button_click(value), bg=button_colors["numbers"]
    )
  button.grid(row=row, column=col)

# Run the application
root.mainloop()


# Group Members
# HS/2020/0159
#Â HS/2020/0421
```
